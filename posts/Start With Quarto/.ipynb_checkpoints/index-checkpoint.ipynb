This is a simple tutorial of how we can read in an online dataset and create a simple visualizaion for Palmer Penguins dataset.

# Data import

We need to first load in the dataset from online repository. We can use `pandas` to do this.

```python
import pandas as pd
# retrieve from online repository
url = "https://raw.githubusercontent.com/PhilChodrow/PIC16B/master/datasets/palmer_penguins.csv"
penguins = pd.read_csv(url)
```

Here we successfully read in the dataset, and then we can start building a simple visualization.

# Plotting

We will be using `matplotlib` for plotting since it offers us great freedom in customizing the graph.

```python
import matplotlib.pyplot as plt

# Create a scatterplot to visualize the relationship between Flipper length and culmen Length
plt.scatter(x = penguins['Flipper Length (mm)'], y = penguins['Culmen Length (mm)'])
# Add a title to the graph
plt.title("Flipper Length VS Culmen Length")
# Name the axises
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Culmen Length (mm)")
plt.show()
```

This is a scatter plot based on palmer penguins dataset. It plots the penguins' flipper length against culmen length. In general, it shows a positive correlation between the two length.
